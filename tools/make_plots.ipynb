This notebook provides an example of how to use the `make_plot` function from `viz.py` to create Bland-Altman, Correlation, and Error-Reference plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import dill as pickle

from viz import *
```

```python
# specify results file
dataset_name = 'mimic'
filter_name = 'hasandazeh19'
alg_name = 'hasandazeh19'

# automated code
results_path = '../results/training/' + dataset_name + '_' + filter_name + '_' + alg_name + '.pickle'
with open(results_path, 'rb') as f:
    data = pickle.load(f)
sbp_model_results = data['sbp']
dbp_model_results = data['dbp']
opt = data['opt']
opt.dataset_name = dataset_name
opt.filter_name = filter_name
opt.alg_name = alg_name

df = make_plot(sbp_model_results, dbp_model_results, 'BA', opt)
```
